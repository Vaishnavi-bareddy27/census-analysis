```python
df = pd.read_csv("population.csv",thousands=",")
print(df.head())
```

```python
df["Decadal Growth (%)"]=df["Decadal Growth (%)"].str.replace("%","").astype(float)
print(df)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df["Literacy Rate (%)"],df["% of Total Population"])
plt.title("Scatter Plot : literacy Rate vs Total Population")
plt.xlabel("Literacy Rate(%)")
plt.ylabel("Total Population")
plt.grid(linestyle="--",alpha=0.5)
plt.savefig("scatter plot.png")
plt.show
```

```python
plt.figure(figsize=(10,10))
plt.bar(df["Literacy Rate (%)"],df["Population"], color="blue",edgecolor="red")
plt.title("BarPlot : literacy Rate vs Population")
plt.ylabel("Literacy Rate(%)")
plt.xlabel("Population")
plt.grid(linestyle="--",alpha=0.5)
plt.savefig("bar plot.png")
plt.show
```

```python
plt.figure(figsize=(10,10))
top_states = df.nlargest(10, "Population")
plt.pie(top_states["Population"],labels=top_states["State"])
plt.title("PiePlot : Top 10 States by Population")
plt.savefig("pie plot.png")
plt.show
```

```python
top_states = df.nlargest(10, "% of Total Population")
x = np.arange(len(top_states["State"]))
width = 0.35
plt.figure(figsize=(12,6))
plt.bar(x - width/2, top_states["Males"], width, label="Males", color="blue")
plt.bar(x + width/2, top_states["Females"], width, label="Females", color="pink")


plt.title("BarPlot :Males vs Females")
plt.savefig("bar plot.png")
plt.show
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df["Urban Population"],df["Rural Population"])
plt.title("Scatter Plot : Urban Population vs Rural Population")
plt.xlabel("Urban Population")
plt.ylabel("Rural Population")
plt.grid(linestyle="--",alpha=0.5)
plt.savefig("scatter plot.png")
plt.show
```
